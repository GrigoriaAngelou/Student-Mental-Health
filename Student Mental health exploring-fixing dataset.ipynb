There is only one missing value in the column of Age. So, I will drop this row because one value will not influence my outcomes.

```python
df=df.dropna()
```

```python
df.isna().sum()
```

The name of the columns are a bit to long, so it will be helpful to rename them with something shorter.

```python
#renaming column names

df.rename(columns = {'Choose your gender':'Gender',
                      'What is your course?':'Course',
                      'Your current year of Study':'Year',
                      'What is your CGPA?':'CGPA',
                      'Do you have Depression?':'Depression',
                      'Do you have Anxiety?':'Anxiety',
                      'Do you have Panic attack?':'Panic attack',
                      'Did you seek any specialist for a treatment?':'Specialist Treatment'}, inplace = True)
df.head()
```

I will not use the Timestamp column so, I will delete it.

```python
del df['Timestamp']
df
```

I can notice that the column Year has different values like year 2 and Year 2 that are both have the same meaning. 
Therefore, I will check all the unique values in every column to see if there is problem with them and I will try
to fix them. Also, I will get rid of the point zero in the Age column (I will change the type from float to integer.

```python
df['Gender'].unique()
```

```python
df['Age'].unique()
```

```python
df['Age'] = df['Age'].astype(int)
df
```

```python
df_year
```

```python
df = df.join(df_year[1])
```
